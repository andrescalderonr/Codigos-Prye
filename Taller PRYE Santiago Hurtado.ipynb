```python
import numpy as np
import matplotlib.pyplot as plt
##Punto 1

#1.	Se utiliza cierto polímero para los sistemas de evacuación de los aviones. Es importante que el polímero sea resistente al proceso de envejecimiento.\n
# Se utilizaron 20 especímenes del polímero en un experimento. 10 Se asignaron aleatoriamente para exponerse a un proceso de envejecimiento acelerado del lote\n
# el cual implica la exposición a altas temperaturas durante 10 días.\n
# se hicieron las mediciones de resistencia a la tensión de los especímenes y se registraron los siguientes datos sobre resistencia a la tensión en psi.

## Sin envejecimiento: 227, 222, 218, 217, 225, 218, 216, 129, 228, 221

## Con envejecimiento: 219, 214, 215, 211, 209, 218, 203, 204, 201, 205

# Datos de resistencia a la tensión en psi:
polimero_sin_envejecimiento = np.array([227,222,218,217,225,218,216,129,228,221])

polimero_con_envejecimiento = np.array([219,214,215,211,209,218,203,204,201,205])

polimero_sin_envejecimiento = np.sort(polimero_sin_envejecimiento)
polimero_con_envejecimiento = np.sort(polimero_con_envejecimiento) 







```

```python
## a) Elabore la gráfica de puntos de los datos

plt.figure(figsize=(10, 6))
plt.plot(polimero_sin_envejecimiento, 'o', label='Sin envejecimiento')
plt.plot(polimero_con_envejecimiento, 'o', label='Con envejecimiento')

# Añadir etiquetas y título
plt.xlabel('Índice')
plt.ylabel('Resistencia a la tensión (psi)')
plt.title('Gráfica de puntos de resistencia a la tensión del polímero')
plt.legend()

# Mostrar la gráfica
plt.show()
```

```python
## c)	Calcule la resistencia a la tensión de la media de la muestra en las dos muestras.

media_sin_envejecimiento = np.mean(polimero_sin_envejecimiento)

media_con_envejecimiento = np.mean(polimero_con_envejecimiento)

print("Resistencia a la tensión de la media del polímero sin envejecimiento:", media_sin_envejecimiento,"psi")
print("Resistencia a la tensión de la media del polímero con envejecimiento:", media_con_envejecimiento,"psi")
```

```python
## d) Calcule la mediana de ambas. Analice la similitud o falta de similitud entre la media y la mediana de cada grupo.

mediana_sin_envejecimiento = np.median(polimero_sin_envejecimiento)
mediana_con_envejecimiento = np.median(polimero_con_envejecimiento)

print("Mediana del polímero sin envejecimiento:", mediana_sin_envejecimiento,"psi")
print("Mediana del polímero con envejecimiento:", mediana_con_envejecimiento,"psi")
```

```python
import math
## e) Calcule la varianza de la muestra y su desviación estándar en cuanto a la resistencia a la tensión en ambas muestras

varianza_sin_envejecimiento = np.var(polimero_sin_envejecimiento, ddof=1) ##se utiliza el parametro ddof = 1 para manejar los grados de libertad
varianza_con_envejecimiento = np.var(polimero_con_envejecimiento, ddof=1)

desviacion_estandar_sin_envejecimiento = math.sqrt(varianza_sin_envejecimiento)
desviacion_estandar_con_envejecimiento = math.sqrt(varianza_con_envejecimiento)

print("varianza de los polimeros sin envejecimiento:", varianza_sin_envejecimiento)
print("varianza de los polimeros con envejecimiento:", varianza_con_envejecimiento)

print("desviación estandar de los polimeros sin envejecimiento", desviacion_estandar_sin_envejecimiento,"psi")
print("desviación estandar de los polimeros con envejecimiento", desviacion_estandar_con_envejecimiento,"psi")
```

```python
## 2.Con base en los datos recolectados en la tabla que se realizó en Excel realice los siguientes puntos. 

## b)	Realice un diagrama de caja y bigote con los datos numéricos

alturas = np.array([1.54, 1.55, 1.58, 1.75, 1.75, 1.87, 1.73, 1.84, 1.79, 1.78, 1.83, 1.70, 1.69, 1.59, 1.68, 1.64, 1.69 , 1.57, 1.74, 1.75, 1.75, 1.65, 1.54, 1.61, 1.74,1.56 ])

alturas = np.sort(alturas)

edades = np.array([18, 20, 20, 19, 20, 21, 18, 21, 18, 19, 20, 20, 22, 23, 19, 19, 19, 19, 22, 20, 21, 19, 19, 22, 20])

edades = np.sort(edades)

tiempos_de_llegada = np.array([20, 45, 10, 90, 70, 90, 30, 30, 20, 90, 50, 90, 30, 90, 40, 20, 40, 90, 30, 25, 35, 90, 39, 110, 40, 90])

tiempos_de_llegada = np.sort(tiempos_de_llegada)



# Diagrama de caja para alturas
plt.figure(figsize=(10, 6))
plt.boxplot(alturas, patch_artist=True, boxprops=dict(facecolor="lightblue"))
plt.title('Diagrama de Caja: Alturas', fontsize=16)
plt.ylabel('Altura (m)', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Diagrama de caja para edades
plt.figure(figsize=(10, 6))
plt.boxplot(edades, patch_artist=True, boxprops=dict(facecolor="lightgreen"))
plt.title('Diagrama de Caja: Edades', fontsize=16)
plt.ylabel('Edad (años)', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Diagrama de caja para tiempos de llegada
plt.figure(figsize=(10, 6))
plt.boxplot(tiempos_de_llegada, patch_artist=True, boxprops=dict(facecolor="lightcoral"))
plt.title('Diagrama de Caja: Tiempos de Llegada', fontsize=16)
plt.ylabel('Tiempo (minutos)', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```

```python
## c)	Realice un histograma con los datos numéricos continuos.

# Histograma para Alturas
plt.figure(figsize=(10, 5))
plt.hist(alturas, bins=8, color='skyblue', edgecolor='black', alpha=0.7)
plt.title("Histograma de Alturas")
plt.xlabel("Altura (m)")
plt.ylabel("Frecuencia")
plt.grid(axis='y', alpha=0.75)
plt.show()


# Histograma para Tiempos de llegada
plt.figure(figsize=(10, 5))
plt.hist(tiempos_de_llegada, bins=8, color='salmon', edgecolor='black', alpha=0.7)
plt.title("Histograma de Tiempos de Llegada")
plt.xlabel("Tiempo de llegada (minutos)")
plt.ylabel("Frecuencia")
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
## e)	Halle la media y mediana para los datos numéricos.
#medias
media_altura = np.mean(alturas)
media_edades = np.mean(edades)
media_tiempos = np.mean(tiempos_de_llegada)

#medianas

mediana_alturas = np.median(alturas)
mediana_edades = np.median(edades)
mediana_tiempos = np.median(tiempos_de_llegada)

print("Media de alturas:", media_altura)
print("Mediana de alturas:", mediana_alturas)

print("Media de edades:", media_edades)
print("Mediana de edades:", mediana_edades)

print("Media de tiempos de llegada:", media_tiempos)
print("Mediana de tiempos de llegada:", mediana_tiempos)
```

```python
## f)	Halle las medidas de variabilidad de los datos numéricos.

#varianzas

varianza_alturas = np.var(alturas, ddof=1 )
varianza_edades = np.var(edades, ddof=1)
varianza_tiempos = np.var(tiempos_de_llegada, ddof=1)

#desviación estandar

desviacion_alturas = math.sqrt(varianza_alturas)
desviacion_edades = math.sqrt(varianza_edades)
desviacion_tiempos = math.sqrt(varianza_tiempos)

print("Varianza de alturas:", varianza_alturas)
print("Desviación estándar de alturas:", desviacion_alturas)

print("Varianza de edades:", varianza_edades)
print("Desviación estándar de edades:", desviacion_edades)

print("Varianza de tiempos de llegada:", varianza_tiempos)
print("Desviación estándar de tiempos de llegada:", desviacion_tiempos)
```

```python
## g)	¿Cuáles son los cuartiles de los datos?

#cuartiles
cuartiles_alturas = np.percentile(alturas, [25, 50, 75])
cuartiles_edades = np.percentile(edades, [25, 50, 75])
cuartiles_tiempos = np.percentile(tiempos_de_llegada, [25, 50, 75])


print("Cuartiles de alturas:", cuartiles_alturas)
print("Cuartiles de edades:", cuartiles_edades)
print("Cuartiles de tiempos de llegada:", cuartiles_tiempos)

```
